### ทำการแยกเดือนและปีออกเป็นคอลัมน์ month และ year


```python
# prompt: การใช้ DataFrame dataset: เพิ่มตารางเดือนและปี

# แยกเดือนและปีออกจากคอลัมน์ 'month'
dataset[['Month', 'Year']] = dataset['month'].str.split('_', expand=True)

# แสดง DataFrame ที่มีคอลัมน์เดือนและปีเพิ่มเข้ามา
dataset = dataset.drop('month', axis=1) # เอา column เก่าออก
dataset
```

### ข้อที่ 1 ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
# prompt: การใช้ DataFrame dataset: หาความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

# แยกข้อมูลตามปี
dataset_2018 = dataset[dataset['Year'].str.contains('2018')]
dataset_2019 = dataset[dataset['Year'].str.contains('2019')]

# คำนวณค่าเฉลี่ยของ Chlo_a สำหรับแต่ละ station ในปี 2018
mean_chlo_a_2018 = dataset_2018.groupby('station')['Chlo_a'].mean()

# คำนวณค่าเฉลี่ยของ Chlo_a สำหรับแต่ละ station ในปี 2019
mean_chlo_a_2019 = dataset_2019.groupby('station')['Chlo_a'].mean()

# หาความแตกต่างของค่าเฉลี่ยระหว่างปี 2018 และ 2019
difference = mean_chlo_a_2018 - mean_chlo_a_2019

print(difference)
```

### ข้อที่ 2 สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
# prompt: การใช้ DataFrame dataset: สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน ผลลัพธ์เป็นตาราง

# คำนวณค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน
result = dataset.groupby(['Month', 'Year','station'])['Chlo_b'].mean().reset_index()

# แสดงผลลัพธ์เป็นตาราง
print(result)
```

```python
import os

dr_path = os.path.join(path,'Data_Docter_ศ.นพ.ชวลิต.csv')
print(dr_path)
```

```python
post = pd.read_csv(dr_path)
post
```

```python
# prompt: การใช้ DataFrame post: ลบ comment ที่ไม่มีคำว่า เป็นมะเร็ง ออก

# Remove rows that don't contain the word เป็นมะเร็ง in the comments column
post_drop = post[post['comments'].str.contains('เป็นมะเร็ง', na=False)]
```

```python
post_drop
```

```python
# List of cancer types
cancer_types = [
    "มะเร็งเต้านม",
    "มะเร็งตับ",
    "มะเร็งต่อมลูกหมาก",
    "มะเร็งปอด",
    "มะเร็งลำไส้",
    "มะเร็งหลอดอาหาร",
    "มะเร็งกระดูก",
    "มะเร็งสมอง",
    "มะเร็งลำไส้ใหญ่",
    "มะเร็งรังไข่",
    "มะเร็งหลังโพรงจมูก",
    "มะเร็งตับอ่อน",
    "มะเร็งระบบอวัยวะสืบพันธุ์สตรี",
    "มะเร็งต่อมน้ำเหลือง",
    "มะเร็งต่อมไทรอยด์",
    "มะเร็งเม็ดเลือดขาว",
    "มะเร็งเม็ดเลือดขาวชนิดเฉียบพลัน",
    "มะเร็งกระเพาะอาหาร",
    "มะเร็งศีรษะและลำคอ",
    "มะเร็งไขกระดูกมัยอิโลมา",
    "มะเร็งท่อทางเดินน้ำดี",
    "มะเร็งผิวหนัง"
]

# Define the function using the list
def cancer_type(comment):
    for cancer in cancer_types:
        if cancer in comment:
            return cancer
    return "มะเร็งชนิดอื่น"

# Apply the function to the DataFrame
post_drop['cancer_type'] = post_drop['comments'].apply(cancer_type)
post_drop['cancer_type'].value_counts()
```

```python
# เปลี่ยนชื่อคอลัมน์เป็นภาษาอังกฤษเพื่อให้ง่ายต่อการจัดการ
post_drop = post_drop.rename(columns={'ยอดไลค์': 'likes', 'จำนวนการตอบกลับ': 'replies'})

# จุดที่น่าสนใจ
# 1. หาโพสต์ที่ได้รับความนิยมสูงสุด 3 โพสต์
most_liked_posts = post_drop.sort_values(by='likes', ascending=False).head(3)
print("Top 3 most liked posts:")
for index, row in most_liked_posts.iterrows():
    print(f"{row['name']} has {row['likes']} likes.")

# 2. หาโรคมะเร็งที่ถูกพูดถึงมากที่สุด 3 ชนิด
top_3_cancer_types = post_drop['cancer_type'].value_counts().head(3)
print("\nTop 3 most common cancer types mentioned in comments:")
print(top_3_cancer_types)
```

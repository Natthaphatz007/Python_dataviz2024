### à¸—à¸³à¸à¸²à¸£à¹à¸¢à¸à¹€à¸”à¸·à¸­à¸™à¹à¸¥à¸°à¸›à¸µà¸­à¸­à¸à¹€à¸›à¹‡à¸™à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ month à¹à¸¥à¸° year


```python
# prompt: à¸à¸²à¸£à¹ƒà¸Šà¹‰ DataFrame dataset: à¹€à¸žà¸´à¹ˆà¸¡à¸•à¸²à¸£à¸²à¸‡à¹€à¸”à¸·à¸­à¸™à¹à¸¥à¸°à¸›à¸µ

# à¹à¸¢à¸à¹€à¸”à¸·à¸­à¸™à¹à¸¥à¸°à¸›à¸µà¸­à¸­à¸à¸ˆà¸²à¸à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'month'
dataset[['Month', 'Year']] = dataset['month'].str.split('_', expand=True)

# à¹à¸ªà¸”à¸‡ DataFrame à¸—à¸µà¹ˆà¸¡à¸µà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¹€à¸”à¸·à¸­à¸™à¹à¸¥à¸°à¸›à¸µà¹€à¸žà¸´à¹ˆà¸¡à¹€à¸‚à¹‰à¸²à¸¡à¸²
dataset = dataset.drop('month', axis=1) # à¹€à¸­à¸² column à¹€à¸à¹ˆà¸²à¸­à¸­à¸
dataset
```

### à¸‚à¹‰à¸­à¸—à¸µà¹ˆ 1 à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡à¸‚à¸­à¸‡ à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ à¸‚à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Chlo_a à¸—à¸¸à¸à¹† station à¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸›à¸µ 2018 à¹à¸¥à¸° à¸›à¸µ 2019

```python
# prompt: à¸à¸²à¸£à¹ƒà¸Šà¹‰ DataFrame dataset: à¸«à¸²à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡à¸‚à¸­à¸‡ à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ à¸‚à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Chlo_a à¸—à¸¸à¸à¹† station à¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸›à¸µ 2018 à¹à¸¥à¸° à¸›à¸µ 2019

# à¹à¸¢à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸•à¸²à¸¡à¸›à¸µ
dataset_2018 = dataset[dataset['Year'].str.contains('2018')]
dataset_2019 = dataset[dataset['Year'].str.contains('2019')]

# à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡ Chlo_a à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸° station à¹ƒà¸™à¸›à¸µ 2018
mean_chlo_a_2018 = dataset_2018.groupby('station')['Chlo_a'].mean()

# à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡ Chlo_a à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸° station à¹ƒà¸™à¸›à¸µ 2019
mean_chlo_a_2019 = dataset_2019.groupby('station')['Chlo_a'].mean()

# à¸«à¸²à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡à¸‚à¸­à¸‡à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸›à¸µ 2018 à¹à¸¥à¸° 2019
difference = mean_chlo_a_2018 - mean_chlo_a_2019

print(difference)
```

### à¸‚à¹‰à¸­à¸—à¸µà¹ˆ 2 à¸ªà¸£à¹‰à¸²à¸‡à¸•à¸²à¸£à¸²à¸‡à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ Chlo_b à¸—à¸¸à¸à¹† station à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸°à¹€à¸”à¸·à¸­à¸™

```python
# prompt: à¸à¸²à¸£à¹ƒà¸Šà¹‰ DataFrame dataset: à¸ªà¸£à¹‰à¸²à¸‡à¸•à¸²à¸£à¸²à¸‡à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ Chlo_b à¸—à¸¸à¸à¹† station à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸°à¹€à¸”à¸·à¸­à¸™ à¸œà¸¥à¸¥à¸±à¸žà¸˜à¹Œà¹€à¸›à¹‡à¸™à¸•à¸²à¸£à¸²à¸‡

# à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢ Chlo_b à¸—à¸¸à¸à¹† station à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸°à¹€à¸”à¸·à¸­à¸™
result = dataset.groupby(['Month', 'Year','station'])['Chlo_b'].mean().reset_index()

# à¹à¸ªà¸”à¸‡à¸œà¸¥à¸¥à¸±à¸žà¸˜à¹Œà¹€à¸›à¹‡à¸™à¸•à¸²à¸£à¸²à¸‡
print(result)
```

```python
# prompt: à¸à¸²à¸£à¹ƒà¸Šà¹‰ DataFrame post: à¸¥à¸š comment à¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¸¡à¸µà¸„à¸³à¸§à¹ˆà¸² à¹€à¸›à¹‡à¸™à¸¡à¸°à¹€à¸£à¹‡à¸‡ à¸­à¸­à¸

# Remove rows that don't contain the word à¹€à¸›à¹‡à¸™à¸¡à¸°à¹€à¸£à¹‡à¸‡ in the comments column
post_drop = post[post['comments'].str.contains('à¹€à¸›à¹‡à¸™à¸¡à¸°à¹€à¸£à¹‡à¸‡', na=False)]
```

```python
# List of cancer types
cancer_types = [
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¹€à¸•à¹‰à¸²à¸™à¸¡",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸•à¸±à¸š",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸•à¹ˆà¸­à¸¡à¸¥à¸¹à¸à¸«à¸¡à¸²à¸",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸›à¸­à¸”",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸¥à¸³à¹„à¸ªà¹‰",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸«à¸¥à¸­à¸”à¸­à¸²à¸«à¸²à¸£",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸à¸£à¸°à¸”à¸¹à¸",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸ªà¸¡à¸­à¸‡",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸¥à¸³à¹„à¸ªà¹‰à¹ƒà¸«à¸à¹ˆ",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸£à¸±à¸‡à¹„à¸‚à¹ˆ",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸«à¸¥à¸±à¸‡à¹‚à¸žà¸£à¸‡à¸ˆà¸¡à¸¹à¸",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸•à¸±à¸šà¸­à¹ˆà¸­à¸™",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸£à¸°à¸šà¸šà¸­à¸§à¸±à¸¢à¸§à¸°à¸ªà¸·à¸šà¸žà¸±à¸™à¸˜à¸¸à¹Œà¸ªà¸•à¸£à¸µ",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸•à¹ˆà¸­à¸¡à¸™à¹‰à¸³à¹€à¸«à¸¥à¸·à¸­à¸‡",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸•à¹ˆà¸­à¸¡à¹„à¸—à¸£à¸­à¸¢à¸”à¹Œ",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¹€à¸¡à¹‡à¸”à¹€à¸¥à¸·à¸­à¸”à¸‚à¸²à¸§",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¹€à¸¡à¹‡à¸”à¹€à¸¥à¸·à¸­à¸”à¸‚à¸²à¸§à¸Šà¸™à¸´à¸”à¹€à¸‰à¸µà¸¢à¸šà¸žà¸¥à¸±à¸™",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸à¸£à¸°à¹€à¸žà¸²à¸°à¸­à¸²à¸«à¸²à¸£",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸¨à¸µà¸£à¸©à¸°à¹à¸¥à¸°à¸¥à¸³à¸„à¸­",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¹„à¸‚à¸à¸£à¸°à¸”à¸¹à¸à¸¡à¸±à¸¢à¸­à¸´à¹‚à¸¥à¸¡à¸²",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸—à¹ˆà¸­à¸—à¸²à¸‡à¹€à¸”à¸´à¸™à¸™à¹‰à¸³à¸”à¸µ",
    "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸œà¸´à¸§à¸«à¸™à¸±à¸‡"
]

# Define the function using the list
def cancer_type(comment):
    for cancer in cancer_types:
        if cancer in comment:
            return cancer
    return "à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸Šà¸™à¸´à¸”à¸­à¸·à¹ˆà¸™"

# Apply the function to the DataFrame
post_drop['cancer_type'] = post_drop['comments'].apply(cancer_type)
post_drop['cancer_type'].value_counts()
```

```python
# à¹€à¸›à¸¥à¸µà¹ˆà¸¢à¸™à¸Šà¸·à¹ˆà¸­à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¹€à¸›à¹‡à¸™à¸ à¸²à¸©à¸²à¸­à¸±à¸‡à¸à¸¤à¸©à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¸‡à¹ˆà¸²à¸¢à¸•à¹ˆà¸­à¸à¸²à¸£à¸ˆà¸±à¸”à¸à¸²à¸£
post_drop = post_drop.rename(columns={'à¸¢à¸­à¸”à¹„à¸¥à¸„à¹Œ': 'likes', 'à¸ˆà¸³à¸™à¸§à¸™à¸à¸²à¸£à¸•à¸­à¸šà¸à¸¥à¸±à¸š': 'replies'})

# à¸ˆà¸¸à¸”à¸—à¸µà¹ˆà¸™à¹ˆà¸²à¸ªà¸™à¹ƒà¸ˆ
# 1. à¸«à¸²à¹‚à¸žà¸ªà¸•à¹Œà¸—à¸µà¹ˆà¹„à¸”à¹‰à¸£à¸±à¸šà¸„à¸§à¸²à¸¡à¸™à¸´à¸¢à¸¡à¸ªà¸¹à¸‡à¸ªà¸¸à¸” 3 à¹‚à¸žà¸ªà¸•à¹Œ
most_liked_posts = post_drop.sort_values(by='likes', ascending=False).head(3)
print("Top 3 most liked posts:")
for index, row in most_liked_posts.iterrows():
    print(f"{row['name']} has {row['likes']} likes.")

# 2. à¸«à¸²à¹‚à¸£à¸„à¸¡à¸°à¹€à¸£à¹‡à¸‡à¸—à¸µà¹ˆà¸–à¸¹à¸à¸žà¸¹à¸”à¸–à¸¶à¸‡à¸¡à¸²à¸à¸—à¸µà¹ˆà¸ªà¸¸à¸” 3 à¸Šà¸™à¸´à¸”
top_3_cancer_types = post_drop['cancer_type'].value_counts().head(3)
print("\nTop 3 most common cancer types mentioned in comments:")
print(top_3_cancer_types)
```

วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน

คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุด

````
mom_dad_count = 0
mom_dad_titles = []
mom_life_count = 0
mom_local_count = 0
creator_counts = {}

for index, row in this_data.iterrows():
    title = row['title']
    if 'แม่' in title or 'พ่อ' in title:
        mom_dad_count += 1
        mom_dad_titles.append(title)
        if row['type'] == 'นิทานชีวิต':
            mom_life_count += 1
        elif row['type'] == 'นิทานประจำถิ่น':
            mom_local_count += 1

        creator = row['creator']
        if creator in creator_counts:
            creator_counts[creator] += 1
        else:
            creator_counts[creator] = 1

print("There are", mom_dad_count, "stories with 'แม่' or 'พ่อ' in the title.")
print("The titles are:", mom_dad_titles)

if mom_life_count > mom_local_count:
    print("There are more stories with 'แม่' or 'พ่อ' in the title in the 'นิทานชีวิต' category.")
elif mom_local_count > mom_life_count:
    print("There are more stories with 'แม่' or 'พ่อ' in the title in the 'นิทานประจำถิ่น' category.")
else:
    print("The number of stories with 'แม่' or 'พ่อ' in the title is the same in both categories.")

most_frequent_creator = max(creator_counts, key=creator_counts.get)
print("The creator who most frequently writes stories with 'แม่' or 'พ่อ' in the title is:", most_frequent_creator)
````
